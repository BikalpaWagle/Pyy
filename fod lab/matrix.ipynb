## Basic Matrix Operations

1.Matrix Operation Perform the following operations on the given matrices: A=[[2 -1], [​0 3​]] B=[[1 ​ 2], [-3 4​]] a. A+B b. AB c. 3A−2B

```python
import numpy as np
```

```python
A = np.array([[2, -1],
              [0,  3]])

B = np.array([[ 1,  2],
              [-3,  4]])
```

```python
#additon of array A and B i.e A+B
c = A + B
print(c)
```

```python
#multication of A and B i.e A*B
AB = A @ B
print(AB)
```

```python
#3A−2B
result = 3*A - 2*B
result
```

## DETERMINANT

```python
import numpy.linalg as LA
```

```python
det_A = LA.det(A)
print(det_A)
```

```python
#since determinant is 6 meaning its inverse exist so
A_inv = LA.inv(A)
print(A_inv)
```

4.You are given a dataset of cars with the following features:Number of cylinders (x1​)Engine size in liters (x2​)Horsepower (x3) car 1: x1: 4 x2: 1.5 x3: 100 and price: 20
car 2: x1: 6 x2: 2.0 x3: 150 and price: 30

Then, find the price of a car 3 with features x1: 8 x2: 3.0 x3: 200 price:?

Hints:

1.Represent the given dataset as a matrix equation X.T*β=y, where X is the feature matrix, β is the coefficient vector, and y is the price vector
2.Compute β (the coefficients) using the normal equation: β=(X.T * X)^−1 X.T y
where, X.T --> transpose matrix

```python
X = np.array([[4, 1.5, 100],
              [6, 2.0, 150]])

y = np.array([[20],
              [30]])

XT_X = X.T @ X
print(XT_X)
G=LA.det(XT_X)
print("DETERMINANT is=",G)
```

-Determinant is 0

(XᵀX) is not invertible
Price of car 3 cannot be uniquely predicted

```python
#pinv is pseudo-inverse which is done when a normal matrix inverse does not exist
beta_pinv = LA.pinv(X) @ y
print(beta_pinv)

x3 = np.array([[8, 3.0, 200]])
price_car3 = x3 @ beta_pinv
print("price of car",price_car3)
```
