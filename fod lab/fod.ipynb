## BAR PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt

# read the csv file
df = pd.read_csv("Health_heart_experimental.csv")

# group by age and calculate average heart rate
avg_hr = df.groupby("age")["HR"].mean()

# plot
plt.figure(figsize=(10,5))
plt.bar(avg_hr.index, avg_hr.values)
plt.xlabel("Age")
plt.ylabel("Average Heart Rate")
plt.title("Average Heart Rate by Age")
plt.show()
```

The bar plot shows the relationship between age and average heart rate. The x-axis represents age, and the y-axis represents the average heart rate. From the graph, it can be seen that the heart rate remains almost constant across different age groups. There are small increases and decreases at some ages, but no major change is observed overall. This indicates that age does not have a strong effect on heart rate in the given data. Overall, the heart rate stays around the same level for people of different ages.

## PIE CHART


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Health_heart_experimental.csv")

# Create age groups
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']

data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)

# Calculate average weight for each age group
age_weight_grouped = data.groupby('age_group')['weightKg'].mean()

#  Remove NaN values (IMPORTANT FIX)
age_weight_grouped = age_weight_grouped.dropna()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_weight_grouped,
        labels=age_weight_grouped.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Age vs Weight (in kg)")
plt.show()
```

The pie chart shows the comparison between different age groups and their average weight in kilograms. Each portion pie represents an age group, and its size indicates the proportion of average weight for that group.The pie chart is almost equally distributed which tells us that the people of different age groups are distributed almost equally on the basis of their weight.This shows that average weight changes gradually with age and does not vary sharply between different age groups.

## BOX PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Health_heart_experimental.csv")

# Prepare data for box plot
bp_data = [data['SysBP'], data['DiaBP']]

# Create box plot
plt.figure(figsize=(8, 6))
plt.boxplot(bp_data, labels=['Systolic BP', 'Diastolic BP'])

# Add title and labels
plt.title("Systolic BP vs Diastolic BP")
plt.ylabel("Blood Pressure (mmHg)")

# Show plot
plt.show()
```

The box plot compares systolic blood pressure and diastolic blood pressure values. From the graph, it is clear that systolic blood pressure values are generally higher than diastolic blood pressure values.The median of systolic BP is higher and from the fig it's range is slighgtly wider so we can say that there is more variation in systolic BP than in diastolic BP. Whereas diastolic Bp has comparatively lower median and small spread range which tells us that there is more variation in systolic BP than in diastolic BP.

### HISTOGRAM


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file from the same folder
df = pd.read_csv("Health_heart_experimental.csv")

# Remove rows with missing age or heart rate
df = df.dropna(subset=["age", "HR"])

# Create age groups
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ["<20", "2029", "3039", "4049", "5059", "60+"]

df["AgeGroup"] = pd.cut(df["age"], bins=bins, labels=labels)

# Plot histogram of heart rate by age group
plt.figure()

for group in labels:
    hr_values = df[df["AgeGroup"] == group]["HR"]
    plt.hist(hr_values, bins=10, alpha=0.5, label=group)

plt.xlabel("Heart Rate (HR)")
plt.ylabel("Frequency")
plt.title("Heart Rate Distribution by Age Group")
plt.legend()

plt.show()

```

The histogram shows how heart rate (HR) values are distributed across different age groups. On the x-axis we have heart rate, and on the y-axis we have frequency.
From the graph, we can see that most heart rate values for all age groups lie between 80 and 160 beats per minute. The bars for different age groups overlap which means that heart rate values are similar across ages.Younger age groups tend to have slightly higher and more spread-out heart rate values, while older age groups show a slightly more range. However, there isn't huge difference between the heart rate of younger age group and the older age group.

### Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file from the same folder
df = pd.read_csv("Health_heart_experimental.csv")

# Remove rows with missing BMI or HR values
df = df.dropna(subset=["BMI", "HR"])

# Create scatter plot
plt.figure()
plt.scatter(df["BMI"], df["HR"])

# Label axes and title
plt.xlabel("Body Mass Index (BMI)")
plt.ylabel("Heart Rate (HR)")
plt.title("Scatter Plot of BMI vs Heart Rate")

# Show plot
plt.show()
```

The above figure shows scatterplot between BMI (body mass index) vs heart rate.BMI is on the x axis whereas heart rate is on the y axis. Higher numbers in BMI means greater weight and higher number on the heart rate means faster heart beat. They both are directly proportional as when BMI increase then the heart rate increases and when the BMI decreases the heart rate also decreases.From the plot people with BMI at the range of 20-25 have a normal heart rate whereas increase in BMI to 30 results in increase in heart rate.
